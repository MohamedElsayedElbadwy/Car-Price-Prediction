# Preprocessing

```python
df
```

```python
from sklearn.preprocessing import OneHotEncoder

le=OneHotEncoder(sparse_output=False,drop='first')

catgorical_features=["fueltype","aspiration","doornumber","carbody","drivewheel","enginelocation","enginetype","cylindernumber","fuelsystem"]

for i in catgorical_features:
    df[i]=le.fit_transform(df[[i]])
```

```python
df
```

```python
x=df.drop(['price','CarName','car_ID'],axis=1)
y=df[['price']]
```

```python
df.info()
```

# Model

```python
from sklearn.model_selection import train_test_split 

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

```

```python
x_train
```

```python
from sklearn.preprocessing import StandardScaler

# hna gm3t kol alnumeric data 34an ts8r kol alarkam
numeric_columns=['symboling','compressionratio','horsepower','peakrpm','citympg','fueltype','aspiration','doornumber','carbody','drivewheel','enginelocation','wheelbase','carlength','carwidth','carheight','curbweight','enginetype' ,'cylindernumber','enginesize','fuelsystem','boreratio','stroke','highwaympg']  

stander_scaler = StandardScaler()
x_train=stander_scaler.fit_transform(x_train[numeric_columns]) # dah kda 3mly ts8er le alarkam
```

```python
x_test=stander_scaler.transform(x_test[numeric_columns])
```

```python
x_test # we hna nfs alklam 
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
```

```python
Model=LinearRegression()
```

```python
Model.fit(x_train,y_train)
```

```python
ypred=Model.predict(x_test)
```

```python
y_test
```

```python
ypred
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
MSE=mean_squared_error(y_test,ypred)
print("MSE Is " , MSE)
```

```python
r2=r2_score(y_test,ypred)
print("R2 Score Is " , r2)
```

```python
plt.figure(figsize=(12,8))
plt.scatter(y_test,ypred) # hna 34an a7ot alpoints 3la almo3adla 

plt.plot([y_test.min(),y_test.max()],[y_test.min(),y_test.max()],color='red')  # deh 34an arsm ben alpoints al line 3la hasb almo3adla

plt.xlabel("Actual Prices")
plt.ylabel("Predict Prices")
plt.title("Predict Prices For Cars")
```
